# ****Dropping rows containing null values and the column 'Per_Sqft'

```python
df_new=df.drop(columns=['Per_Sqft'])
df_drop=df_new.dropna()
df_new.isnull().sum()
```

# Applying target encoding on locality

```python
target_encoding = df_drop.groupby('Locality')['Price'].mean()

# Replace 'locality' with the mean house price for that locality
df_drop['Locality_encoded'] = df_drop['Locality'].map(target_encoding)

print(df_drop)
```

```python
from sklearn.model_selection import train_test_split
X=df_drop.drop(columns=['Price'])
y=df_drop.iloc[:,6]
X.sample(5)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

# Applying ordinal encoding on furinshing

```python
from sklearn.preprocessing import OrdinalEncoder

encoder=OrdinalEncoder(categories=[['Unfurnished','Semi-Furnished','Furnished']])

X_train['Furnishing']=encoder.fit_transform(X_train[['Furnishing']])
X_test['Furnishing']=encoder.transform(X_test[['Furnishing']])
```

```python
X_train
```

# Applying one hot encoding on 'Status','Transaction','Type'

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

column_transformer = ColumnTransformer(
    transformers=[
        ('Onehot', OneHotEncoder(sparse=False,drop='first'), ['Status','Transaction','Type']),
    ],
    remainder='passthrough' 
)
```

```python
X_train
```

```python
X_train1=X_train
```

```python
X_train_transformed=column_transformer.fit_transform(X_train1)
X_test_transformed=column_transformer.transform(X_test)
X_train_transformed
```

# Dropping first column of 'Status','Transaction','Type' after applying one hot encoding

```python
X_train_transformed=pd.DataFrame(X_train_transformed,columns=['Status','Transaction','Type','Area','BHK','Bathroom','Furnishing','Locality','Parking','Locality_encoded'])
X_test_transformed=pd.DataFrame(X_test_transformed,columns=['Status','Transaction','Type','Area','BHK','Bathroom','Furnishing','Locality','Parking','Locality_encoded'])
X_test_transformed
```

```python
X_train_transformed_drop_locality=X_train_transformed.drop(columns=['Locality'])
X_test_transformed_drop_locality=X_test_transformed.drop(columns=['Locality'])
X_test_transformed_drop_locality
```

# Not dropping first column of 'Status','Transaction','Type' after one hot encoding

```python
column_transformer = ColumnTransformer(
    transformers=[
        ('Onehot', OneHotEncoder(sparse=False), ['Status','Transaction','Type']),
    ],
    remainder='passthrough'  # Leave other columns (like 'Name') unchanged
)

X_train_transformed2=column_transformer.fit_transform(X_train1)
X_test_transformed2=column_transformer.transform(X_test)
X_train_transformed2.shape

X_train_transformed2=pd.DataFrame(X_train_transformed2,columns=['Status_rtm','Status_ar','Transaction_resale','Transaction_new_property','Type_apartment','Type_building_floor','Area','BHK','Bathroom','Furnishing','Locality','Parking','Locality_encoded'])
X_test_transformed2=pd.DataFrame(X_test_transformed2,columns=['Status_rtm','Status_ar','Transaction_resale','Transaction_new_property','Type_apartment','Type_building_floor','Area','BHK','Bathroom','Furnishing','Locality','Parking','Locality_encoded'])
X_test_transformed2
```

```python
X_train_transformed2_drop_locality=X_train_transformed2.drop(columns=['Locality'])
X_test_transformed2_drop_locality=X_test_transformed2.drop(columns=['Locality'])
```

# Training the model using random forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
rfr=RandomForestRegressor(n_estimators=100,bootstrap=True,min_samples_split=3)

rfr.fit(X_train_transformed2_drop_locality,y_train)

y_pred1=rfr.predict(X_test_transformed2_drop_locality)
print(r2_score(y_test,y_pred1))
```

# Using Decision tree

```python
from sklearn.tree import DecisionTreeRegressor

dtr=DecisionTreeRegressor(min_samples_leaf=2)
dtr.fit(X_train_transformed2_drop_locality,y_train)

y_pred2=dtr.predict(X_test_transformed2_drop_locality)
print(r2_score(y_test,y_pred2))
```

# Using linear regressor

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train_transformed_drop_locality,y_train)

y_pred3=lr.predict(X_test_transformed_drop_locality)
print(r2_score(y_test,y_pred3))
```

# Using gradient boosting

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(n_estimators=150, learning_rate=0.05, max_depth=4, random_state=42)
gbr.fit(X_train_transformed2_drop_locality,y_train)

y_pred4=gbr.predict(X_test_transformed2_drop_locality)
print(r2_score(y_test,y_pred4))
```

# Using grid search cv

```python
from sklearn.model_selection import GridSearchCV

# Define a parameter grid to search for best hyperparameters
param_grid = {
    'n_estimators': [100,150, 200, 300],
    'learning_rate': [0.01, 0.1, 0.05],
    'max_depth': [3, 4, 5]
}

# Initialize the GridSearch with cross-validation
grid_search = GridSearchCV(GradientBoostingRegressor(random_state=42), param_grid, cv=5, n_jobs=-1,scoring='r2')

# Fit the model with different hyperparameters
grid_search.fit(X_train_transformed2_drop_locality, y_train)

# Best parameters found
print("Best parameters: ", grid_search.best_params_)
```
